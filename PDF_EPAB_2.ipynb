```python
# @title 📚 PDF to EPUB Converter (нажмите ▶ для запуска)

# Установка необходимых пакетов
!pip install -q pdfminer.six ebooklib

from google.colab import files
import os
from pdfminer.high_level import extract_text
from ebooklib import epub

# Загрузка PDF
uploaded = files.upload()
if not uploaded:
    print("❌ Файл не загружен!")
else:
    pdf_name = next(iter(uploaded))
    epub_name = pdf_name.replace('.pdf', '.epub')

    try:
        # Конвертация
        print("⏳ Идет конвертация...")
        text = extract_text(pdf_name)

        # Создание EPUB
        book = epub.EpubBook()
        book.set_identifier('id'+pdf_name)
        book.set_title(os.path.splitext(pdf_name)[0])
        book.set_language('ru')
        book.add_author('Converted with Colab')

        # Добавляем главу с текстом
        chapter = epub.EpubHtml(title='Content', file_name='chap_01.xhtml', lang='ru')
        chapter.content = f'<h1>{book.title}</h1><pre>{text}</pre>'
        book.add_item(chapter)

        # Структура книги
        book.toc = (epub.Link('chap_01.xhtml', 'Содержание', 'intro'),)
        book.spine = ['nav', chapter]
        book.add_item(epub.EpubNcx())
        book.add_item(epub.EpubNav())

        # Сохранение
        epub.write_epub(epub_name, book, {})
        print(f'✅ Готово! Файл "{epub_name}" создан')

        # Скачивание
        files.download(epub_name)

    except Exception as e:
        print(f'❌ Ошибка: {e}')
        print('Попробуйте другой PDF-файл')
```
